Abaixo temos um dataframe com as seguintes informações:


**Nome:** O nome de cada indivíduo da base


**Idade:** A idade de cada um deles


**Profissão:** A atual ocupação


**Salário:** O salário registrado


**Limite de crédito:** O limite de crédito disponível para cada um dos indivíduos


**Historico_Inadimplencia:** Marco como 0 aqueles que nunca tiveram um histórico de inadimplência e como 1 aqueles que já tiveram em algum momento.


**Estado civil:** O estado civil de cada um deles


**Imovel_Proprio:** Marcado como 0 aqueles que não tem imóvel próprio e como 1 aqueles que tem.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python

data = {
    'Nome': ['Alice', 'Joao', 'Charlie', 'David', 'Eva', 'Diego', 'Denize', 'Claudio'],
    'Idade': [25, 30, 35, 40, 45, 60, 22, 24],
    'Profissão': ['Engenheiro', 'Médico', 'Professor', 'Advogado', 'Médico','Engenheiro', 'Estudante','Estudante'],
    'Salário': ['4500', '8000', '5000', '10000', '12000','15000', '1200','1500'],
    'Limite_Credito': ['2500', '4000', '4000', '1000', '10000','2000', '500','250'],
    'Historico_Inadimplencia': ['0', '0', '0', '1', '0','1', '0','1'],
    'Estado_Civil': ['Casamento', 'Casamento', 'Solteiro', 'Solteiro', 'Casamento','Solteiro', 'Solteiro','Solteiro'],
    'Imovel_Proprio': ['0', '0', '0', '1', '1','1', '0','0']
}

df = pd.DataFrame(data)
df.head()

# tratamento de dados
df['Salário'] = df['Salário'].astype(float)
df['Limite_Credito'] = df['Limite_Credito'].astype(float)
df['Historico_Inadimplencia'] = df['Historico_Inadimplencia'].astype(int)
df['Imovel_Proprio'] = df['Imovel_Proprio'].astype(int)

# criando variavel de limite alto
media_limite = df['Limite_Credito'].mean()
df['limite_alto'] = df['Limite_Credito'] > media_limite
```

# 📊 Gráfico 1
## Salário vs Limite de Crédito

### 🎯 Objetivo

Identificar se salário é o principal fator impulsionador do limite de crédito.

---
### 🔎 Evidência nos dados

| Salário | Limite |
| ------- | ------ |
| 12000   | 10000  |
| 8000    | 4000   |
| 5000    | 4000   |
| 1500    | 250    |

---
### 📌 Conclusão

Existe uma relação fortemente proporcional entre salário e limite de crédito.

---
### 📈 Tipo de gráfico ideal

Scatter Plot (Salário vs Limite)

---
### 💡 Mensagem transmitida pelo gráfico

Quanto maior o salário, maior tende a ser o limite de crédito.

👉 Este é provavelmente o principal fator impulsionador.


```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Salário', y='Limite_Credito')
plt.title('Salário vs Limite de Crédito')
plt.xlabel('Salário')
plt.ylabel('Limite de Crédito')
plt.show()
```

# 📊 Gráfico 2
## Média do Limite por Profissão


---
### 🎯 Objetivo

Ver se profissão influencia diretamente no limite de crédito.

---
### 🔎 Observação nos dados

- Médicos → 8000 salário / 4000 limite

- Médicos → 12000 salário / 10000 limite

- Professor → 5000 salário / 4000 limite

- Engenheiros → comportamento misto

---
### 📌 Conclusão

Profissões com maior renda média (principalmente Médico) apresentam maior limite de crédito.

---
### 📊 Tipo de gráfico ideal

Barplot (média do limite por profissão)

---
### 💡 Mensagem transmitida pelo gráfico

Profissões com maior renda tendem a receber maiores limites de crédito.

```python
plt.figure(figsize=(10, 6))
media_profissao = df.groupby('Profissão')['Limite_Credito'].mean().reset_index()
sns.barplot(data=media_profissao, x='Profissão', y='Limite_Credito')
plt.title('Média do Limite de Crédito por Profissão')
plt.xlabel('Profissão')
plt.ylabel('Média do Limite')
plt.xticks(rotation=45)
plt.show()
```
